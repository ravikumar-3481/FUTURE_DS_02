# Click Through Rate (CTR)
-  Percentage of impressions that resulted in clicks
- Let's calculate CTR to know  how engaging each ad was relative to impressions

```python
df['CTR'] = (df['clicks'] / df['impressions']) * 100
df['CTR'].head()
```

```python
df[['ad_id', 'impressions', 'clicks', 'CTR']].sort_values(by='CTR', ascending=False).head(10)
```

# CPC

```python
import numpy as np

df['CPC'] = np.where(df['clicks'] > 0, df['spent'] / df['clicks'], 0)
df['CPC'].head()
```

# CPA

```python
# Calculate CPA per ad 
df['CPA'] = np.where(df['approved_conversion'] > 0,df['spent'] / df['approved_conversion'],0) 

# Overall CPA (campaign level)
overall_cpa = df['spent'].sum() / df['approved_conversion'].sum()

print("Overall CPA:", round(overall_cpa, 2))

# Preview results
df[['ad_id', 'spent', 'approved_conversion', 'CPA']].head(10)
```

# Conversion Rate

```python
df['Conversion Rate'] = np.where(df['clicks'] > 0,(df['approved_conversion'] / df['clicks']) * 100,0)
df['Conversion Rate'].head()
```

# ROI

```python
df['ROI'] = np.where(df['spent'] > 0,(df['approved_conversion'] * 100) / df['spent'],0)
df['ROI'].head()
```

# 3. What was the Click-Through Rate (CTR) and Return on Investment (ROI)?

```python
# Overall CTR
overall_ctr = (df['clicks'].sum() / df['impressions'].sum()) * 100

# Overall ROI (using approved conversions as "return")
overall_roi = (df['approved_conversion'].sum() * 100) / df['spent'].sum()

print(f"Overall CTR: {overall_ctr:.2f}%")
print(f"Overall ROI: {overall_roi:.2f}")
```

#  Interpretation of Campaign Performance

**Overall CTR: 0.01%**  
- This means that out of every 10,000 ad impressions, only about 1 person clicked.  
- The CTR is extremely low, suggesting that the ads did not strongly capture audience attention or that the targeting/creative wasn’t engaging enough.  

**Overall ROI: 2.98**  
- This indicates that for every **1 dollar spent**, the campaign generated the equivalent of **2.98 dollar in approved conversions**.  
- Despite the low engagement (CTR), the campaign delivered a **high return on investment**, meaning that the few people who did engage were highly valuable and likely converted at a strong rate.  

---

## Key Takeaways
- Ads may not be grabbing attention broadly (**low CTR**), but they are very effective at converting the small group who do engage (**high ROI**).  
- Future improvements could focus on **increasing CTR** (better creatives, targeting, or messaging) to **scale results while maintaining profitability**.

# CTR By Age Group

```python
import matplotlib.pyplot as plt
import seaborn as sns

# First calculate CTR by age group
ctr_age = df.groupby("age")['CTR'].mean().reset_index()

# Now get colors based on number of age groups
colors = sns.color_palette("Set1", len(ctr_age))

# Plot
plt.figure(figsize=(8,5))
plt.bar(ctr_age['age'], ctr_age['CTR'], color=colors)
plt.title("CTR by Age Group")
plt.xlabel("Age Group")
plt.ylabel("CTR (%)")
plt.xticks(rotation=45)
plt.show()
```

# ROI By Gender

```python
roi_gender = df.groupby("gender")['ROI'].mean().reset_index()

# Use Pastel1 palette
colors = sns.color_palette("viridis", len(roi_gender))

plt.figure(figsize=(6,5))
plt.bar(roi_gender['gender'], roi_gender['ROI'], color=colors)
plt.title("ROI by Gender")
plt.xlabel("Gender")
plt.ylabel("ROI")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(
    data=df,
    x="spent",
    y="approved_conversion",
    hue="age",         
    size="clicks",      
    sizes=(200,800),    
    alpha=0.7,
    palette="tab10",    
    edgecolor="k",
    legend="full" )    

# Remove the unwanted size legend (keep only age)
handles, labels = plt.gca().get_legend_handles_labels()
    
# First part = ages, second part = sizes ... keep only ages
n_age = df['age'].nunique()
plt.legend(handles[:n_age], labels[:n_age], title="Age", loc='upper right')

plt.title("Spend vs Approved Conversions", fontsize=14, weight="bold")
plt.xlabel("Spend ($)")
plt.ylabel("Approved Conversions")
plt.tight_layout()
plt.show()
```
